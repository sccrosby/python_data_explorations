## Try to predict cancellation
Using lead_time, arrival_date_month, adults, children, babies, previous cancellations, previous_bookings_not_canceled

```python
# select and remove NaN rows
X = df[['lead_time', 'adults', 'children', 'babies', 'previous_cancellations', 'previous_bookings_not_canceled','is_canceled','arrival_date_month']]
X = X.dropna(axis=0)
X.shape

# select y
y = X['is_canceled']

# Drop y from X
X = X.drop('is_canceled',axis=1)

# select month
X_month = X['arrival_date_month']
X = X.drop('arrival_date_month',axis=1)

# Convert to numpy
X = X.to_numpy()
```

```python
y.shape
```

```python
X.shape
```

```python
# Do one-hot encoding for month
from sklearn.preprocessing import OneHotEncoder
enc = OneHotEncoder(handle_unknown='ignore')

X_enc = X_month.to_numpy().reshape(-1, 1)

enc.fit(X_enc)
X_enc = enc.transform(X_enc)
```

```python
print(X.shape)
print(X_enc.shape)
```

```python
Xm = np.concatenate((X,X_enc.todense()),axis=1)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import cross_val_score, train_test_split
from sklearn.pipeline import Pipeline

X_train, X_test, y_train, y_test = train_test_split(Xm,y)

pipeline = Pipeline([
    ("scalar", StandardScaler()),
    ("log_reg", LogisticRegression())
])

cross_val_score(pipeline, X_train, y_train, cv=3, scoring="accuracy")

```

```python
scalar = StandardScaler()
X_train_tr = scalar.fit_transform(X_train)
X_test_tr = scalar.transform(X_test)

clf_log = LogisticRegression()
clf_log.fit(X_train_tr,y_train)

pred = clf_log.predict(X_test_tr)

from sklearn.metrics import precision_score, recall_score
ps = precision_score(y_test,pred)
rs = recall_score(y_test,pred)

print(ps, rs)

# Of identified cancellations, 70% are correct.  (precision)
# Of all true cancellations, 30% are identified. (recall)
```

```python
clf_log.coef_
clf_log.classes_
```

```python
from sklearn.model_selection import cross_val_predict
from sklearn.metrics import precision_recall_curve

y_scores = cross_val_predict(clf_log, X_train_tr, y_train, cv=3, method='decision_function')

precisions, recalls, thresholds = precision_recall_curve(y_train, y_scores)

def plot_precision_recall_vs_threshold(precisions, recalls, thresholds):
    plt.plot(thresholds, precisions[:-1], 'b--', label='Precision')
    plt.plot(thresholds, recalls[:-1], 'g-',label='Recall')
    plt.legend()

plot_precision_recall_vs_threshold(precisions,recalls,thresholds)
plt.grid()
plt.xlim((-2,2))
plt.show()
```
